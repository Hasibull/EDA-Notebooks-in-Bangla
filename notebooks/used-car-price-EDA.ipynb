```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

বেশির ভাগ ক্ষেত্রেই ডেটা csv (comma separated value) ফরমেটেই থাকে, সেক্ষেত্রে ডেটা read করার জন্য pandas এর read_csv() ফাংশনটি ব্যাবহার করা হয়। তবে ডেটা যদি অন্য কোন ফরমেটে থেকে থাকে তাহলে তা read করার জন্য pandas এ আরও কিছু ফাংশন রয়েছে। যেমন,

- CSV file - read_csv()
- JSON file - read_json()
- Excel file - read_excel()
- Database table - read_sql()

```python
df = pd.read_csv('../datasets/used_car_price/train_data.csv')
```

**head()** ফাংশনটি ডিফল্ট ভাবে শুরুর ৫ টি row প্রদর্শন করে। তবে আমরা চাইলে কতগুলো row প্রদর্শন করতে চাই তা মেনশন করে দিতে পারি। যেমন, নিচে ৬ টি row প্রদর্শন করা হয়েছে।

```python
df.head(6)
```

**tail()** ফাংশনটি ডিফল্ট ভাবে শেষের ৫ টি row প্রদর্শন করে। তবে আমরা চাইলে কতগুলো row প্রদর্শন করতে চাই তা মেনশন করে দিতে পারি। যেমন, নিচে ৩ টি row প্রদর্শন করা হয়েছে।

ডেটাসেটের প্রতিটি column এ কি টাইপের ডেটা রয়েছে এবং প্রতিটি column এ কতগুলো ডেটা non-null ডেটা রয়েছে তা জানার জন্য আমরা **info()** ফাংশন ব্যাবহার করতে পারি!

```python
df.info()
```

ডেটাসেটের প্রতিটি column এ কতগুলো ডেটা null বা missing রয়েছে তা জানার জন্য **isnull()** ফাংশনটি অনেক কার্যকরী। তবে শেষে **sum()** ফাংশন যুক্ত করতে ভুল করা যাবে না!!

```python
df.isnull().sum()
```
